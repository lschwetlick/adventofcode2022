```python
food_log = {}
elfnr = 1
food_log[elfnr] = []
with open("input.txt", "r") as file:
    for line in file:
        line = line.strip()
        if line == "":
          elfnr += 1
          food_log[elfnr] = []
        else:
          food_log[elfnr].append(int(line))
food_log
```

# solution 1

```python
biggest_sum = 0
for elfnr in food_log:
  if (s:=sum(food_log[elfnr])) > biggest_sum:
      biggest_sum = s
biggest_sum
```

# solution 2

```python
sums = []
for elfnr in food_log:
   sums.append(sum(food_log[elfnr]))
sums.sort()
print(sums[-3:])
sum(sums[-3:])
```
